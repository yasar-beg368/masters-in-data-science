Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Ans:  Matplotlib is a popular Python library used for creating static, animated, and interactive visualizations in Python. It provides a wide range of functions and classes for generating high-quality 2D and 3D plots and figures. Matplotlib is widely used in various scientific, engineering, and data analysis fields to visualize data and explore patterns and relationships.

Here are five plots that can be plotted using the Pyplot module of Matplotlib:

Line Plot: A line plot is a basic plot where data points are connected with straight lines. It is commonly used to display the trend or relationship between two variables over a continuous interval.


Scatter Plot: A scatter plot displays individual data points as markers, with each point representing the value of two different variables. Scatter plots are useful for visualizing the distribution and relationship between two variables.

Bar Plot: A bar plot (or bar chart) represents data using rectangular bars, where the length or height of each bar corresponds to the value it represents. Bar plots are commonly used to compare categories or groups of data.

Histogram: A histogram represents the distribution of a continuous numerical variable by dividing it into bins and displaying the frequency or count of data points falling into each bin. It provides insights into the shape, spread, and central tendency of the data.

Pie Chart: A pie chart is a circular chart divided into sectors, where each sector represents a category or a portion of a whole. Pie charts are useful for showing proportions and percentages of categorical data.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

```python
#ans:  A scatter plot is a type of plot that displays individual data points as markers on a two-dimensional coordinate system. 
#It is commonly used to visualize the relationship between two continuous variables and identify patterns, trends, or correlations.



import numpy as np
import matplotlib.pyplot as plt
x=np.random.rand(10)
y=np.random.rand(10)
plt.scatter(x,y)
plt.xlabel("x_values",c='r')
plt.ylabel("y_values",c='r')
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
ans:  """The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. 
It allows you to display multiple plots in separate axes, arranged in rows and columns, within a single figure window. 
This function is particularly useful when you want to compare or visualize multiple plots side by side"""



import numpy as np
import matplotlib.pyplot as plt

# Data for line plots
x = np.array([0, 1, 2, 3, 4, 5])

# Line 1
y1 = np.array([0, 100, 200, 300, 400, 500])

# Line 2
y2 = np.array([50, 20, 40, 20, 60, 70])

# Line 3
y3 = np.array([10, 20, 30, 40, 50, 60])

# Line 4
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Line 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Line 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Line 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Line 4')

# Adjust spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

```python
Ans:  """A bar plot, also known as a bar chart, is a type of plot that presents categorical data with rectangular bars. 
The length or height of each bar corresponds to the quantity or frequency of the data it represents.
Bar plots are commonly used to compare and display the relative values of different categories or groups.

Bar plots are useful for several reasons:

Comparison: Bar plots allow easy visual comparison of data across different categories. They help identify variations, trends, and differences in data values.

Categorical Data: Bar plots are particularly suitable for visualizing categorical or discrete data, where each category is represented by a bar.

Presentation: Bar plots provide an effective way to present data in a visually appealing and easy-to-understand manner. They are widely used in reports, presentations, and publications to convey information clearly """



import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.xlabel("company")
plt.ylabel("profit")
plt.show()

plt.barh(company,profit)
plt.xlabel("company")
plt.ylabel("profit")
plt.show()
```
