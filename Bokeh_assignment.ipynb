Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.


In Bokeh, glyphs are visual elements that represent data points in a plot. They can be used to create various types of markers, lines, bars, and other graphical representations of data.

To add glyphs to a Bokeh plot, you typically follow these steps:
Create a figure: Use the figure function from bokeh.plotting to create a new figure object that will hold your plot.

Add glyphs to the figure: Use glyph methods such as line, circle, rect, or patch to add specific types of glyphs to the figure. These methods take data and other parameters to determine the appearance and position of the glyphs.

Customize the glyphs: You can customize the appearance of glyphs by specifying additional parameters such as line color, fill color, size, or transparency.

Display or save the plot: Finally, you can use the show function from bokeh.io to display the plot in a browser or the save function to save it as an HTML file.

```python
from bokeh.plotting import figure, show

# Prepare your data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a figure
p = figure(title="My Bokeh Plot", x_axis_label="X", y_axis_label="Y")

# Add a line glyph
p.line(x, y, line_color="blue", line_width=2)

# Add circle glyphs
p.circle(x, y, fill_color="red", size=10)

# Display the plot
show(p)
```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

Bokeh provides a range of options to customize the appearance of a plot, including axes, title, and legend. Here's an overview of how you can customize these elements:

Axes customization:

Axis labels: You can set the labels of the x-axis and y-axis using the x_axis_label and y_axis_label properties of the figure object.
Axis range: You can control the range of values displayed on the axes using the x_range and y_range properties of the figure object.
Axis ticks: You can customize the appearance of axis ticks, including their location, labels, and style, using various properties such as x_ticks, x_minor_ticks, y_ticks, y_minor_ticks, etc.
Title customization:

Plot title: You can set the title of the plot using the title property of the figure object.
Title text properties: You can customize the appearance of the title text, such as font size, color, alignment, etc., using the title_text_font_size, title_text_color, title_text_align, and other related properties.
Legend customization:

Legend placement: You can control the placement of the legend on the plot using the legend.location property of the figure object.
Legend labels: You can set the labels for each glyph in the legend using the legend.label property of the glyph method. For example, p.line(x, y, legend_label="Line 1").
Legend appearance: You can customize the appearance of the legend, such as font size, color, background color, etc., using properties like legend.label_text_font_size, legend.label_text_color, legend.background_fill_color, and others.

```python
from bokeh.plotting import figure, show

# Prepare your data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a figure
p = figure(title="My Bokeh Plot", x_axis_label="X", y_axis_label="Y",
           x_range=(0, 6), y_range=(0, 12))

# Customize axes ticks
p.xaxis.ticker = [1, 2, 3, 4, 5]
p.yaxis.ticker = [0, 5, 10]

# Customize title text
p.title.text_font_size = "16pt"
p.title.text_color = "navy"

# Add line and circle glyphs with legend labels
p.line(x, y, line_color="blue", line_width=2, legend_label="Line 1")
p.circle(x, y, fill_color="red", size=10, legend_label="Circles")

# Customize legend appearance
p.legend.location = "top_right"
p.legend.label_text_font_size = "12pt"
p.legend.background_fill_color = "lightgray"

# Display the plot
show(p)
```

Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?

To embed a Bokeh plot into a web page or dashboard using Flask or Django, you need to follow these general steps:

Using Flask:

Set up Flask: Start by setting up a Flask application. This involves creating a Flask project, setting up routes, and configuring the Flask environment.

Create a Bokeh plot: Use the Bokeh library to create the desired plot. Define the glyphs, customize the plot, and set up any necessary callbacks.

Save the Bokeh plot as an HTML file: Use the save function from bokeh.io to save the Bokeh plot as an HTML file. Provide a filename for the HTML file.

Serve the HTML file in Flask: In your Flask route, render the HTML file and pass it to the template engine to be displayed.

Display the Bokeh plot in your web page: In the HTML template of your Flask route, insert the Bokeh plot's HTML file using the appropriate template syntax.

Using Django:

Set up Django: Start by setting up a Django project. Create a Django app within the project and configure the Django environment.

Create a Bokeh plot: Use the Bokeh library to create the desired plot. Define the glyphs, customize the plot, and set up any necessary callbacks.

Save the Bokeh plot as an HTML file: Use the save function from bokeh.io to save the Bokeh plot as an HTML file. Provide a filename for the HTML file.

Serve the HTML file in Django: In your Django view, render the HTML file and pass it as a context variable to be displayed.

Display the Bokeh plot in your web page: In the Django template associated with your view, insert the Bokeh plot's HTML file using the appropriate template syntax.

```python
from flask import Flask, render_template
from bokeh.plotting import figure, save

app = Flask(__name__)

@app.route('/')
def index():
    # Create a Bokeh plot
    p = figure()
    p.line([1, 2, 3], [4, 5, 6])

    # Save the Bokeh plot as an HTML file
    save(p, 'bokeh_plot.html')

    # Render the HTML file in the Flask template
    return render_template('index.html', bokeh_plot='bokeh_plot.html')

if __name__ == '__main__':
    app.run(debug=True)
```
